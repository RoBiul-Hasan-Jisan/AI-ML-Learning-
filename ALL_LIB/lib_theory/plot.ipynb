# Scatter Plot

Use when:

 You want to visualize relationship between two numerical features

 Good for seeing clusters, patterns, outliers

Example:
 Age vs Salary, Height vs Weight

Best use:

 Regression visualization

 Clustering visualization

 Outlier detection

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,5,4,5]

plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Scatter Plot")
plt.show()
```

# Line Plot
Use when:

 Data is sequential or time-series

 To see trend over time

Example:
 Stock price vs time, Loss vs epochs, Accuracy vs epochs

Best use:

 Model training curves (loss/accuracy)

 Time series forecasting

```python
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot")
plt.show()
```

# Histogram
Use when:

 To see distribution of a numerical feature

 To find skewness, spread, normal distribution, outliers
Example:
 Histogram of age, Histogram of pixel intensity

Best use:

 Feature understanding

 Preprocessing decisions (normalize? log transform?)

```python
import numpy as np

data = np.random.randn(1000)  
plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

# Box Plot

Use when:

 To compare distribution across multiple categories

 To detect outliers clearly
Example:
 Salary by gender, Score distribution by class

Best use:

 Outlier detection

 Comparing variance of features across groups

```python
plt.boxplot(data)
plt.title("Box Plot")
plt.show()
```

# Bar Plot

Use when:

 Feature is categorical

 Want to visualize counts / frequency

Example:
Count of classes, How many images per label

Best use:

 Class imbalance check

 Categorical feature exploration

```python
categories = ["Cat A", "Cat B", "Cat C"]
values = [30, 50, 20]

plt.bar(categories, values)
plt.title("Bar Plot")
plt.show()
```

# Heatmap

Use when:

 Want to visualize correlation between multiple features

 Want to see confusion matrix clearly

Example:
 Correlation heatmap, Confusion matrix heatmap

Best use:

 Feature selection

 Understanding model mistakes

```python
import seaborn as sns
import numpy as np

data = np.random.rand(5,5)
sns.heatmap(data, annot=True)
plt.title("Heatmap")
plt.show()
```

# Pair Plot (Seaborn)

Use when:

 You want to visualize all feature relationships at once

 Best for small feature datasets (38 features)

Example:
 Iris dataset pair plot

Best use:

 Quick data understanding

 Seeing clusters and patterns

```python
import seaborn as sns
from seaborn import load_dataset

df = load_dataset("iris")
sns.pairplot(df, hue="species")
plt.show()
```

# Violin Plot

Use when:

 To compare distribution + density across categories

 Shows more detail than a boxplot

Best use:

 Comparing numerical feature across labels (classification)

```python
sns.violinplot(data=df, x="species", y="sepal_length")
plt.title("Violin Plot")
plt.show()
```

# Confusion Matrix Plot
Use when:

 You are evaluating a classification model

 Helps see where model is making mistakes

Example:
 True Positive, False Positive, False Negative, True Negative

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = [0,1,0,1,0,1]
y_pred = [0,1,0,0,1,1]

cm = confusion_matrix(y_true, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.title("Confusion Matrix")
plt.show()
```

# ROC Curve

Use when:

 Evaluating binary classifier

 To check performance at all thresholds

Shows:

 True Positive Rate (TPR)

 False Positive Rate (FPR)

```python
from sklearn.metrics import roc_curve
import numpy as np

# Fake predictions
y_true = [0,1,0,1,0,1]
y_prob = [0.2,0.8,0.3,0.7,0.1,0.6]

fpr, tpr, _ = roc_curve(y_true, y_prob)
plt.plot(fpr, tpr)
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.show()
```

# Precision-Recall Curve

Use when:

 The dataset is imbalanced

 Better than ROC in such cases

```python
from sklearn.metrics import precision_recall_curve

precision, recall, _ = precision_recall_curve(y_true, y_prob)
plt.plot(recall, precision)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.show()
```
