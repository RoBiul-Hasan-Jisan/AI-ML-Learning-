```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 14, 18, 23, 30]

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
plt.plot(x, y, color='red', linestyle='--', marker='o', markersize=8)
plt.title("Customized Line Plot")
plt.show()
```

```python
plt.plot(x, y, label='Growth', color='blue')
plt.plot(x, [i*5 for i in x], label='Double', color='orange')

plt.title("Two Lines with Legend")
plt.xlabel("X-axis")
plt.ylabel("Values")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

ax[0].plot(x, y, color='green')
ax[0].set_title("Plot 1")

ax[1].bar(x, y, color='purple')
ax[1].set_title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, cmap='viridis')
plt.title("2D Contour Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar(label="Z value")
plt.show()
```

```python
plt.contourf(X, Y, Z, cmap='plasma')
plt.title("Filled Contour Plot")
plt.colorbar(label="Z value")
plt.show()
```

```python
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(x, y)
U = -Y
V = X

plt.quiver(X, Y, U, V)
plt.title("2D Vector Field (Quiver Plot)")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

t = np.linspace(0, 10, 100)
x = np.sin(t)
y = np.cos(t)
z = t

ax.plot3D(x, y, z, color='red')
ax.set_title("3D Line Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

ax.scatter3D(x, y, z, color='green', s=50)
ax.set_title("3D Scatter Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Surface Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='black')
ax.set_title("3D Wireframe Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.contour3D(X, Y, Z, 50, cmap='coolwarm')
ax.set_title("3D Contour Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

_x = np.arange(5)
_y = np.arange(5)
_xx, _yy = np.meshgrid(_x, _y)
x, y = _xx.ravel(), _yy.ravel()
z = np.zeros_like(x)
dx = dy = 0.5
dz = np.random.randint(1, 10, len(z))

ax.bar3d(x, y, z, dx, dy, dz, shade=True)
ax.set_title("3D Bar Plot")
plt.show()
```

```python
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(elev=30, azim=45)  # Adjust camera angle
```
